```python
import pandas
import matplotlib.pyplot as plt
df = pandas.read_csv("data/dengue_labels_train.csv")

sj = df[(df.city == "sj")]
sj = sj.drop(columns=['city', 'year', "weekofyear"])
sj.plot()
plt.xlabel("weeks")
plt.show()

iq = df[(df.city == "iq")]
iq = iq.drop(columns=['city', 'year', "weekofyear"])

iq.plot()
plt.xlabel("weeks")
plt.show()
```
