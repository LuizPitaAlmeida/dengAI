```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Both cities train data analysis 

Plot curves without segment cities

```python
x_train = pd.read_csv('../data/dengue_features_train.csv')
x_train.head(10)
```

```python
y_train = pd.read_csv('./data/dengue_labels_train.csv')
y_train.head(10)
```

```python
y_train['total_cases'].plot()
plt.xlabel("weeks")
plt.legend()
plt.show()
```

## San Juan City data analysis

Data from 1990 to 2008

```python
city_sj = y_train[y_train.city == 'sj']
city_sj.head(5)
```

```python
city_sj['total_cases'].plot()
plt.xlabel("weeks")
plt.legend()
plt.show()
```

```python
years = city_sj['year'].unique()
plt.figure(figsize=(20,8))
for year in years:
    temp_sj = city_sj['total_cases'].loc[city_sj['year']==year]
    plt.plot(temp_sj, label = year)
plt.xlabel("weeks")
plt.legend()
plt.show()
```
